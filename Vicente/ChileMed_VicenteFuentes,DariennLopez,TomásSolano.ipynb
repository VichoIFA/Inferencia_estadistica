```python


def datos_KStest(datos_a_analizar):
    
    distribuciones = [
        ('Normal', scipy.stats.norm),
        ('Gamma', scipy.stats.gamma),
        ('Log-normal', scipy.stats.lognorm),
        ('Weibull', scipy.stats.weibull_min),
        ('Exponencial', scipy.stats.expon)
    ]

    datos = []

    for nombre, distribucion in distribuciones:

        parametros = distribucion.fit(datos_a_analizar)
        
        statistic, p_value = scipy.stats.kstest(datos_a_analizar, distribucion.cdf, args=parametros)
        
        datos.append({
            'Distribución': nombre,
            'Parámetros': parametros,
            'KS value': statistic,
            'p-value': p_value
        })
    
    resultados_df = pd.DataFrame(datos)
    #mejor_distro_Pvalue = resultados_df[resultados_df["p-value"] > 0.05]
    #resultados_df = mejor_distro_Pvalue.reset_index().drop("index", axis = "columns")
    #indice = resultados_df["KS value"].idxmin()
    #data_seleccionada = [resultados_df.loc[indice]]
    
    return (pd.DataFrame(resultados_df))

valores = list(df["SERVICIO_SALUD"].unique())
for i in valores:
    filtro = df[df["SERVICIO_SALUD"] == i]
    estancia = filtro["TIEMPO_ESTANCIA_DIAS"].to_list()
    data = datos_KStest(estancia)
    print(i)
    display(data)
    sns.histplot(estancia, kde=True, discrete=True)
    plt.xlim(3)
    plt.show()
    print("================================")
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.stats.multicomp import pairwise_tukeyhsd

df_anova = df[df['SERVICIO_SALUD'].notna() & df['TIEMPO_ESTANCIA_DIAS'].notna()].copy()

modelo = ols('TIEMPO_ESTANCIA_DIAS ~ C(SERVICIO_SALUD)', data=df_anova).fit()
tabla_anova = sm.stats.anova_lm(modelo, typ=2)

print("Análisis de Varianza (ANOVA) - Tiempo de Estancia entre METROPOLITANO y el resto de servicios de salud")
print(tabla_anova)

alpha = 0.05
p_valor = tabla_anova['PR(>F)'][0]

if p_valor < alpha:
    print(f"\nDecisión: Rechazamos la hipótesis nula (p < {alpha})")
    print("Conclusión: Existe al menos una diferencia significativa en el tiempo de estancia medio entre METROPOLITANO y el resto de servicios de salud")
    
    tukey = pairwise_tukeyhsd(endog=df_anova['TIEMPO_ESTANCIA_DIAS'],
                              groups=df_anova['SERVICIO_SALUD'],
                              alpha=0.05)
    print("\nResultados prueba post-hoc de Tukey HSD:")

    resumen_tukey = pd.DataFrame(data=tukey.summary().data[1:], columns=tukey.summary().data[0])
    comparaciones_metropolitano = resumen_tukey[
        (resumen_tukey['group1'] == 'METROPOLITANO') | (resumen_tukey['group2'] == 'METROPOLITANO')
    ]
    print("\nComparaciones de Servicio de salud METROPOLITANO contra el resto:")
    display(comparaciones_metropolitano)
    
else:
    print(f"\nDecisión: No rechazamos la hipótesis nula (p ≥ {alpha})")
    print("Conclusión: No hay evidencia suficiente para afirmar que existen diferencias entre METROPOLITANO y el resto de servicios de salud")

stats_by_servicio = df_anova.groupby('SERVICIO_SALUD')['TIEMPO_ESTANCIA_DIAS'].agg(['count', 'mean', 'std', 'min', 'max'])
print("\nEstadísticas descriptivas del Tiempo de Estancia por Servicio de Salud:")
print(stats_by_servicio)
```
